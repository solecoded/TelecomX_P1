# 📤 [1] Extracción (E - Extract)

**Importa .json / Verifica que la conversión a DataFrame sea exitosa**

```python
import pandas as pd
import requests
import json
import plotly.express as px
from IPython.display import display, Markdown  # Importación corregida

url_api = "https://raw.githubusercontent.com/solecoded/TelecomX_P1/main/data/TelecomX_Data.json"

response = requests.get(url_api)

if response.status_code == 200:
    datos_json = response.json()
    df = pd.DataFrame(datos_json)
    display(Markdown("✅Los datos se han cargados exitosamente en un DataFrame."))
    display(Markdown("## 📑 Primeras 5 filas del DataFrame:"))
    display(df.head())
    display(Markdown("## 📋 Información del DataFrame:"))
    display(df.info())
else:
    display(Markdown(f"⛔ Error al cargar los datos. Código de estado: {response.status_code}"))
```

**Resumen del DataFrame**

```python
df.info()
display(Markdown("## 📑 Tipos de datos de cada columna:"))
display(df.dtypes)
```

# 🔄 [2] Transformación (T - Transform)

**Aplanado de columnas**

```python
df = pd.concat([df, pd.json_normalize(df['customer'])], axis=1)
df = pd.concat([df, pd.json_normalize(df['phone'])], axis=1)
df = pd.concat([df, pd.json_normalize(df['internet'])], axis=1)
df = pd.concat([df, pd.json_normalize(df['account'])], axis=1)

df.drop(columns=['customer', 'phone', 'internet', 'account'], inplace=True)

display(Markdown("## 📋 DataFrame con columnas aplanadas:"))

df.info()
```

**Mapeo nombre de columnas a sus traducciones**

```python
columnas_traducidas = {
    'customerID': 'ID_cliente',
    'Churn': 'churn',
    'customer': 'cliente',
    'phone': 'telefono',
    'internet': 'internet',
    'account': 'cuenta',
    'gender': 'genero',
    'SeniorCitizen': 'adultoMayor',
    'Partner': 'pareja',
    'Dependents': 'dependientes',
    'tenure': 'permanenciaMeses',
    'PhoneService': 'servicioTelefonico',
    'MultipleLines': 'multiplesLineas',
    'InternetService': 'servicioInternet',
    'OnlineSecurity': 'seguridadOnline',
    'OnlineBackup': 'respaldoOnline',
    'DeviceProtection': 'proteccionDispositivo',
    'TechSupport': 'soporteTecnico',
    'StreamingTV': 'streamingTV',
    'StreamingMovies': 'streamingPeliculas',
    'Contract': 'tipoContrato',
    'PaperlessBilling': 'facturacionOnline',
    'PaymentMethod': 'metodoPago',
    'Charges.Monthly': 'cargosMensuales',
    'Charges.Total': 'cargosTotales'
}

df = df.rename(columns=columnas_traducidas)
display(Markdown("## 📋 Nombres nuevos de las columnas"))
display(df.columns)
```

**Normalización columna cargosTotales**

```python
df['cargosTotales'] = pd.to_numeric(df['cargosTotales'], errors='coerce')

display(Markdown("## 📋 Valores nulos después de la conversión"))
display(df.isnull().sum())

df.dropna(subset=['cargosTotales'], inplace=True)

display(Markdown("## ✅ Verificación final"))
df.info()
```

**Normalización situación sentimental**

```python
if 'pareja' in df.columns:
    pareja_mapping = {'No': 0, 'Yes': 1}
    df['pareja'] = df['pareja'].map(pareja_mapping)
    df['pareja'] = df['pareja'].fillna(0).astype(int)

display(Markdown("✅ ¡La columna 'pareja' se ha procesado correctamente!"))
```

**Normalizacion datos de internet**

```python
import pandas as pd
from IPython.display import display, Markdown

binary_service_mapping = {'No': 0, 'Yes': 1, 'No phone service': 0, 'No internet service': 0}
service_cols_to_map = [
    'servicioTelefonico', 'multiplesLineas', 'seguridadOnline', 'respaldoOnline',
    'proteccionDispositivo', 'soporteTecnico', 'streamingTV', 'streamingPeliculas'
]

for col in service_cols_to_map:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = df[col].map(binary_service_mapping)
        df[col] = df[col].fillna(0)
        df[col] = df[col].astype(int)

if 'servicioInternet' in df.columns and df['servicioInternet'].dtype == 'object':
    df['servicioInternet'] = df['servicioInternet'].replace({'No internet service': 'No'})

display(Markdown("✅ ¡Las columnas de servicio se han procesado correctamente!"))
```

**Normalización columna Dependientes**

```python
import pandas as pd
from IPython.display import display, Markdown

if 'dependientes' in df.columns and df['dependientes'].dtype == 'object':
    df['dependientes'] = df['dependientes'].map({'Yes': 1, 'No': 0}).astype(int)
    display(Markdown("✅ Columna 'dependientes' convertida a 0/1."))
elif 'dependientes' in df.columns and df['dependientes'].isin([0, 1]).all():
    display(Markdown("⚠️ **La columna 'dependientes' ya está en formato 0/1.**"))
else:
    display(Markdown("⚠️ **Advertencia:** La columna 'dependientes' no se encuentra o no tiene los valores esperados para mapeo (Yes/No o 0/1)."))

display(Markdown("## 📋 Valores únicos en la columna 'dependientes' después de la estandarización"))
display(df['dependientes'].unique())

display(Markdown("## 📋 Conteo de valores en la columna 'dependientes'"))
display(df['dependientes'].value_counts())
```

**Verificación de inconsistencias en la columna "Churn"**

```python
from IPython.display import display, Markdown

filas_con_churn_invalido = df[~df['churn'].isin(['Yes', 'No'])]

display(Markdown("## ⚠️ Filas con valores inconsistentes en la columna 'Churn':"))
display(filas_con_churn_invalido)
```

**Limpieza y recálculo de la columna 'Churn'**

```python
from IPython.display import display, Markdown

df['churn'] = df['churn'].apply(lambda x: x if x in ['Yes', 'No'] else None)
df.dropna(subset=['churn'], inplace=True)

display(Markdown("## 🕵️‍♂️ Verificación de la limpieza"))
df.info()

tasa_evasion = df['churn'].value_counts(normalize=True) * 100
display(Markdown("## 📊 Nueva tasa de abandono de clientes"))
display(tasa_evasion.map('{:.2f}%'.format))
```

**DataFrame con la nueva columna 'Cuentas Diarias'**

```python
from IPython.display import display, Markdown

df['cuentasDiarias'] = df['cargosMensuales'] / 30

display(Markdown("## 📑 DataFrame con la nueva columna 'Cuentas Diarias'"))
display(df[['cargosMensuales', 'cuentasDiarias']].head())
```

**Transformación de variables categóricas a binarias (0/1)**

```python
from IPython.display import display, Markdown

if 'genero' in df.columns:
    genero_mapping = {'Male': 0, 'Female': 1}
    df['genero'] = df['genero'].map(genero_mapping)
    df['genero'] = df['genero'].fillna(0)
    df['genero'] = df['genero'].astype(int)

display(Markdown("✅ ¡La columna 'genero' se ha procesado correctamente!"))
```

**Mapeo al español de valores en la columna tipoContrato**

```python
from IPython.display import display, Markdown
import pandas as pd

contratos_binario = {
    'Month-to-month': 0,
    'One year': 1,
    'Two year': 2
}

df['tipoContrato'] = df['tipoContrato'].replace(contratos_binario)

display(Markdown("## 📜 Valores únicos en 'tipoContrato' después de mapeo"))
display(df['tipoContrato'].unique())

churn_por_contrato = pd.crosstab(df['tipoContrato'], df['churn'], normalize='index') * 100

display(Markdown("## 📊 Tasa de abandono por tipo de contrato:"))
display(churn_por_contrato.style.format('{:.2f}%'))
```

**Mapeo numérico de valores en la columna metodoPago**

```python
from IPython.display import display, Markdown

metodoPago_numerico = {
    'Electronic check': 0,
    'Mailed check': 1,
    'Bank transfer (automatic)': 2,
    'Credit card (automatic)': 3
}

df['metodoPago'] = df['metodoPago'].replace(metodoPago_numerico)

distribucion_metodo_pago = df['metodoPago'].value_counts(normalize=True) * 100

display(Markdown("## 📊 Distribución de la columna 'metodoPago' después de mapeo:"))
display(distribucion_metodo_pago.map('{:.2f}%'.format))
```

**Mapeo de valores numéricos en la columna servicioInternet**

```python
from IPython.display import display, Markdown

servicioInternet_numerico = {
    'No': 0,
    'Fiber optic': 1,
    'DSL': 2,
}

df['servicioInternet'] = df['servicioInternet'].replace(servicioInternet_numerico)

distribucion_servicioInternet = df['servicioInternet'].value_counts(normalize=True) * 100

display(Markdown("## 📊 Distribución de la columna 'servicioInternet' después de mapeo:"))
display(distribucion_servicioInternet.map('{:.2f}%'.format))
```

**Estandarización de columnas binarias churn y facturacionOnline**

```python
import pandas as pd
from IPython.display import display, Markdown

display(Markdown("## 👷🏻👷🏻‍♀️ Estandarizando columnas 'facturacionOnline' y 'churn' a 0/1"))

if 'facturacionOnline' in df.columns and df['facturacionOnline'].dtype == 'object':
    df['facturacionOnline'] = df['facturacionOnline'].map({'Yes': 1, 'No': 0}).fillna(0).astype(int)
    display(Markdown("✅ Columna 'facturacionOnline' convertida a 0/1."))
elif 'facturacionOnline' in df.columns and df['facturacionOnline'].isin([0, 1]).all():
    display(Markdown("✅ La columna 'facturacionOnline' ya está en formato 0/1."))
else:
    display(Markdown("⚠️ Advertencia: La columna 'facturacionOnline' no se encuentra o no tiene los valores esperados (Yes/No o 0/1)."))

if 'churn' in df.columns and df['churn'].dtype == 'object':
    df['churn'] = df['churn'].map({'Yes': 1, 'No': 0}).fillna(0).astype(int)
    display(Markdown("✅ Columna 'churn' convertida a 0/1."))
elif 'churn' in df.columns and df['churn'].isin([0, 1]).all():
    display(Markdown("✅ La columna 'churn' ya está en formato 0/1."))
else:
    display(Markdown("⚠️ Advertencia: La columna 'churn' no se encuentra o no tiene los valores esperados (Yes/No o 0/1)."))

display(Markdown("## 🔍 Valores únicos en 'facturacionOnline' después de la estandarización"))
display(df['facturacionOnline'].unique())
display(Markdown("## 🔍 Valores únicos en 'churn' después de la estandarización "))
display(df['churn'].unique())
```

📅 **Métricas (media, mediana y desviación estándar) para permanencia, cargos mensuales y cargos totales**

```python
from IPython.display import display, Markdown

variables_numericas = ['permanenciaMeses', 'cargosMensuales', 'cargosTotales']
display(Markdown(f"#📅 Métricas (media, mediana y desviación estándar) para permanencia, cargos mensuales y cargos totales"))

for var in variables_numericas:
    mediana_clientes_no = df[df['churn'] == 0][var].median()
    mediana_clientes_si = df[df['churn'] == 1][var].median()

    display(Markdown(f"### Mediana de **{var}**"))
    display(Markdown(f"* Clientes que no se fueron (`churn` = 0): **{mediana_clientes_no:.2f}**"))
    display(Markdown(f"* Clientes que sí se fueron (`churn` = 1): **{mediana_clientes_si:.2f}**"))

    try:
        moda_clientes_no = df[df['churn'] == 0][var].mode()[0]
        moda_clientes_si = df[df['churn'] == 1][var].mode()[0]
        display(Markdown(f"### Moda de **{var}**"))
        display(Markdown(f"* Clientes que no se fueron (`churn` = 0): **{moda_clientes_no:.2f}**"))
        display(Markdown(f"* Clientes que sí se fueron (`churn` = 1): **{moda_clientes_si:.2f}**"))
    except IndexError:
        display(Markdown(f"***Nota:*** No se encontró una moda única para **{var}**."))
```

Análisis de métricas por Churn

A diferencia del promedio (media), la mediana y la moda son menos sensibles a valores extremos, lo que las hace muy útiles para entender el comportamiento típico de los grupos.

## 🔍 **Hallazgos**

**Permanencia en Meses:** La mediana de permanencia para los clientes que se quedaron (38 meses) es casi cuatro veces mayor que la de los clientes que se fueron (10 meses). Esto confirma que la antigüedad del cliente es el factor más crítico para predecir el abandono. Los clientes que se van lo hacen muy temprano en su ciclo de vida.

**Cargos Mensuales:** La mediana y la moda de los cargos mensuales son significativamente más altas para los clientes que se fueron. 📅 **Análisis de Churn por variables categóricas**

```python
import pandas as pd
from scipy.stats import chi2_contingency
from IPython.display import display, Markdown

mapeo_servicios = {
    'genero': mapeo_genero,
    'adultoMayor': mapeo_adultoMayor,
    'pareja': mapeo_pareja,
    'dependientes': mapeo_dependientes,
    'servicioTelefonico': mapeo_servicioTelefonico,
    'multiplesLineas': mapeo_multiplesLineas,
    'servicioInternet': mapeo_servicioInternet,
    'seguridadOnline': mapeo_seguridadOnline,
    'respaldoOnline': mapeo_respaldoOnline,
    'proteccionDispositivo': mapeo_proteccionDispositivo,
    'soporteTecnico': mapeo_soporteTecnico,
    'streamingTV': mapeo_streamingTV,
    'streamingPeliculas': mapeo_streamingPeliculas,
    'tipoContrato': mapeo_tipoContrato,
    'facturacionOnline': mapeo_facturacionOnline,
    'metodoPago': mapeo_metodoPago
}

variables_categoricas = ['genero', 'adultoMayor', 'pareja', 'dependientes', 'servicioTelefonico',
                         'multiplesLineas', 'servicioInternet', 'seguridadOnline', 'respaldoOnline',
                         'proteccionDispositivo', 'soporteTecnico', 'streamingTV', 'streamingPeliculas',
                         'tipoContrato', 'facturacionOnline', 'metodoPago']

display(Markdown(f"#📅 Análisis de Churn por variables categóricas"))

for var in variables_categoricas:

    display(Markdown(f'### Análisis estadístico para: **{var.capitalize()}**'))

    tabla_contingencia = pd.crosstab(df[var], df['churn'])
    tabla_contingencia = tabla_contingencia.rename(index=mapeo_servicios[var], columns=mapeo_churn)
    display(tabla_contingencia)
    chi2, p_value, dof, expected = chi2_contingency(tabla_contingencia)
    display(Markdown(f'- **🔢 Valor p (p-value):** {p_value:.4f}'))

    if p_value < 0.05:
        display(Markdown(f'- **Conclusión:** ⚠️ Existe una relación estadísticamente significativa entre **{var.capitalize()}** y el abandono (Churn).'))
    else:
        display(Markdown(f'- **Conclusión:** 🔹 No hay evidencia de una relación estadísticamente significativa entre **{var.capitalize()}** y el abandono (Churn).'))

    display(Markdown('<br>'))
    display(Markdown('---'))
```

## 🗂️ **Descripción**

Este análisis estadístico evalúa la relación entre las principales variables categóricas de los clientes y su estado de abandono (Churn). Se utilizó una prueba estadística Chi-cuadrado para determinar si existe una relación significativa entre las variables. El valor p es la métrica clave para esta determinación: un valor p menor a 0.05 indica una relación estadísticamente significativa.

## 🔍 **Hallazgos**

El 27.49% de sus cargos totales se pierden debido al churn, en comparación con solo el 15.42% del grupo de no adultos mayores.
